```python
# !pip install requests
# !pip install beautifulsoup4

import requests
from bs4 import BeautifulSoup
import csv
import random

from google.colab import drive
drive.mount('/drive')

words = []

with open('/drive/My Drive/Colab Notebooks/English_Game/final_EnglishGame_dataset.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
      words.append(row[0])

random.shuffle(words)
words = words[:5]
meanings = []
word_meaning = {}
# words = ["cat","bat","mat","rat","sit"]

for word in words:
  web = requests.get('https://www.wordhippo.com/what-is/the-opposite-of/'+word+'.html')
  # print(word,":")
  data = web.content
  soup = BeautifulSoup(data,features="html.parser")
  # print(soup)
  tag = soup.find_all("div","relatedwords")
  # print(tag)

  for i in tag:
    text = i.text
    text = text[1:].split("\n")
    random.shuffle(text)
    meanings.append(text[0])
    word_meaning[word]=text[0]
    break

# print(meanings)
# print(word_meaning)

user_ans={}
random.shuffle(words)
random.shuffle(meanings)

print("\nMatch the given words with there antonym - ")
print(f'{"Words":{20}} {" ":{5}} Antonyms(Options)')

for i in range(len(words)):
  print(f'{str(chr(65+i))+") "+words[i]:{20}} {"-":{5}} {str(i+1)+") "+meanings[i]}')

print("\nYour answers here - ")
for i in range(len(words)):
  print(f'{str(chr(65+i))+") "+words[i]:{20}} {"-":{5}}',end=" ")
  ans = int(input())
  user_ans[words[i]] = meanings[ans-1]

score = 0
for i in list(word_meaning.keys()):
  if(user_ans[i]==word_meaning[i]):
    score+=1


# print(words)
# print(meanings)
# print(word_meaning)
# print(user_ans)
print("\nScore: ",score,"/",len(user_ans))

print("\nCorrect answers are - ")
for i in list(word_meaning.keys()):
  print(i,"►",word_meaning[i],"\n")
```
